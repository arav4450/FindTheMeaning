```python
with open('text.txt') as f:
    lines = f.readlines()
```

```python
import re
from bs4 import BeautifulSoup

def decontracted(phrase):
    # specific
    phrase = re.sub(r"won't", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)

    # general
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase

def preprocess(sentence):

    x = re.sub(r"http\S+", "", sentence)
    x = BeautifulSoup(x, 'lxml').get_text()
    x = decontracted(x)
    x = re.sub("\S*\d\S*", "", x).strip()
    x= re.sub('[^A-Za-z]+', ' ', x)
    x = ' '.join(e.lower() for e in x.split() )
    return x.strip().split()
```

```python
def build_data(text):
  length = 20
  lines = []
  chunks = [text[i:i+length] for i in range(0, len(text), length)]
  for index, chunk in enumerate(chunks):
    line = ' '.join(word for word in chunks[index])
    lines.append(line)
  return lines
```

```python
txt = preprocess(lines[0])
build_data(txt)
```

```python
data = []
for line in lines:
  line_processed = build_data(preprocess(line))
  data =  data + line_processed


len(data)
```

```python
data_final = { }
data_final["context"] = data
data_final["answers"] = {"answer_start":[],"text":[]}
data_final["question"] = []
```

```python
import random


for item in data_final["context"]:
  words = item.strip().split()
  range = len(words)
  random_no = random.randint(0,range-1)
  word = words[random_no]
  prefix = ' '.join(e for e in words[:random_no])
  if random_no == 0:
    index = 0
  else:
    index = len(prefix) + 1

  data_final["answers"]["answer_start"].append(index)
  data_final["answers"]["text"].append(word)
  data_final["question"].append(word+"?")
```
